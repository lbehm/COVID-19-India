```r

  #### within how many days, cases double
  doublesIN = NULL
  for(today in growth$Day[]){
    if((today == 1) || (growth[which(growth$Day == today), 'Total.Confirmed'] == 1))
      dIN = 0
    else {
      half = (growth[which(growth$Day == today),'Total.Confirmed']/2)
      # 39> <38
      halfDay = growth[which(growth$Total.Confirmed == half), 'Day']
      if(length(halfDay) != 0){
        dIN = today-halfDay
        
      } else {
        lowerBound = max(growth[which(growth$Total.Confirmed <= half), 'Day'])
        upperBound = min(growth[which(growth$Total.Confirmed >= half), 'Day'])
        
        if(length(lowerBound) == 0)
          dIN = upperBound
        else {
          subL = c(half - growth[which(growth$Day == lowerBound), 'Total.Confirmed'])
          subU = c(growth[which(growth$Day == upperBound), 'Total.Confirmed']- half)
          
          dIN = today - ifelse((subL<subU), lowerBound, upperBound)
        }
      }
    }
    
    if(length(dIN) > 1)
      dIN = max(dIN)
    
    doublesIN = c(doublesIN, dIN)
  }
  
  ### adding col 
  growth$"Doubles In" = doublesIN
  
  
  ######## new cases
  day = growth$Day
  cases = growth$Total.Confirmed
  
  newCases = NULL
  for(i in day){
    if(i==1)
      newCases = c(newCases, cases[i])
    else{
      val = cases[i]-cases[i-1]
      newCases = c(newCases, val)
    }
  }
  
  ### adding new
  growth$"New Cases" = newCases
  return(growth)
```

```r
# naming affected people
totalPpl = growth$Total.Confirmed[nrow(growth)]
names = NULL
for(i in 1:totalPpl){
    names = c(names, paste("P",i,sep = ""))
}

## confirmed ---> all P1 to P519, so
Conf = factor(rep(1, 519), levels = c(0,1))

## recovered or not
totalRecr = growth$Recovered[nrow(growth)]
recovered = factor(c(rep(1, totalRecr), rep(0, c(519-totalRecr))), levels = c(0,1))

## dead or not
tatalDead = growth$Deaths[nrow(growth)]
deaths = factor(c(rep(1, tatalDead), rep(0, c(519-tatalDead))), levels = c(0,1))

## active or not
tatalActive = growth$Active.Cases[nrow(growth)]
active = factor(c(rep(1, tatalActive), rep(0, c(519-tatalActive))), levels = c(0,1))

## closed or not
tatalClosed = growth$Closed.Cases[nrow(growth)]
closed = factor(c(rep(1, tatalClosed), rep(0, c(519-tatalClosed))), levels = c(0,1))

# status, 1-Recovered; 2-Dead; 3-Active
status = factor(c(rep(1, totalRecr), rep(2, tatalDead), rep(3, tatalActive)), levels = c(1, 2, 3))
```

```r
#creating a strange dataset
wast = data.frame(
            People = names,
            Status = status,
            Confirmed = Conf,
            Recovered = recovered,
            Deaths = deaths,
            Active = active,
            Closed = closed
            )
## altering 
# rcvr
wast$Recovered[which(wast$Status == 1)] <- 1
wast$Recovered[which(!wast$Status == 1)] <- 0

# dead
wast$Deaths[which(wast$Status == 2)] <- 1
wast$Deaths[which(!wast$Status == 2)] <- 0

# active
wast$Active[which(wast$Status == 3)] <- 1
wast$Active[which(!wast$Status == 3)] <- 0

# rcvr
wast$Closed[which(wast$Active == 0)] <- 1
wast$Closed[which(!wast$Active == 0)] <- 0

#table(wast$Recovered)
#table(wast$Deaths)
#table(wast$Active)
#table(wast$Closed)

wast[30:100,]
#totalRecr + tatalDead + tatalActive
```
